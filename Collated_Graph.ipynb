```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from PIL import Image

# Load the saved images
network_img = Image.open("network_graph.png")
sankey_img = Image.open("sankey_diagram.png")
bar_img = Image.open("bar_chart.png")

# Create a figure with a 2x2 grid layout
fig = plt.figure(figsize=(14, 8.5))  # Long bond paper size (14x8.5 inches)
gs = GridSpec(2, 2, width_ratios=[1, 2], height_ratios=[1, 1])  # 2x2 grid with custom ratios

# Top-left: Bar Chart
ax1 = fig.add_subplot(gs[0, 0])
ax1.imshow(bar_img)
ax1.axis('off')  # Hide axes
ax1.set_title('Bar Chart', fontsize=10, pad=10)

# Bottom-left: Sankey Diagram
ax2 = fig.add_subplot(gs[1, 0])
ax2.imshow(sankey_img)
ax2.axis('off')  # Hide axes
ax2.set_title('Sankey Diagram', fontsize=10, pad=10)

# Right-side (spanning both rows): Network Plot
ax3 = fig.add_subplot(gs[:, 1])  # Span all rows in the second column
ax3.imshow(network_img)
ax3.axis('off')  # Hide axes
ax3.set_title('Network Plot', fontsize=10, pad=10)

# Adjust layout to prevent overlap
plt.tight_layout()

# Save the combined figure
plt.savefig("combined_graphs_2x2.png", dpi=300, bbox_inches='tight')

# Display the combined figure
plt.show()
```
